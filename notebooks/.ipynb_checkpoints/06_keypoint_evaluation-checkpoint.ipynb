# MouserCV Keypoint Detection Evaluation — DLC SuperAnimal-Quadruped on Side-View Cage Video

Evaluates DLC SuperAnimal-Quadruped zero-shot keypoint detection quality
on real Bauer lab videos (EGFR pruritus mouse model). Specifically tests
whether paw keypoints are reliable enough for scratch/groom discrimination.

**Key questions:**
1. Which keypoints are reliably detected in side-view cage footage?
2. Are paw keypoints (front + hind) detected above pcutoff consistently?
3. Can paw velocity differentiate scratching from grooming?
4. Is hind paw elevation visible during scratch bouts?

**Runtime:** Google Colab with T4 GPU, or local with CUDA.

```python
import os

# === CONFIGURATION ===
# Path to video (adjust for Colab: upload or mount GDrive)
VIDEO_PATH = "../data/videos/CV analysis project/Cage 2.MOV"

# Time range to evaluate
START_TIME = "1:00"
END_TIME = "2:00"

# DLC model
SUPERANIMAL_NAME = "superanimal_quadruped"
MODEL_NAME = "hrnet_w32"
DETECTOR_NAME = "fasterrcnn_resnet50_fpn_v2"

# Confidence threshold for "reliable" keypoints
PCUTOFF = 0.3

# Essential keypoints for behavior discrimination
ESSENTIAL_KEYPOINTS = [
    "nose",
    "front_left_paw", "front_right_paw",  # grooming
    "back_left_paw", "back_right_paw",    # scratching
    "tail_base",                           # body orientation
]

# All 39 SuperAnimal-Quadruped keypoints for reference
ALL_KEYPOINTS = [
    "nose", "upper_jaw", "lower_jaw", "mouth_end_right", "mouth_end_left",
    "right_eye", "right_earbase", "right_earend", "right_antler_base", "right_antler_end",
    "left_eye", "left_earbase", "left_earend", "left_antler_base", "left_antler_end",
    "neck_base", "neck_end", "throat_base", "throat_end",
    "back_base", "back_end", "back_middle",
    "tail_base", "tail_end",
    "front_left_thai", "front_left_knee", "front_left_paw",
    "front_right_thigh", "front_right_knee", "front_right_paw",
    "back_left_paw", "back_left_thigh",
    "back_right_thai", "back_left_knee", "back_right_knee", "back_right_paw",
    "belly_bottom", "body_middle_right", "body_middle_left",
]
```

```python
def time_to_seconds(time_str: str) -> float:
    """Convert mm:ss or mm:ss.ms or hh:mm:ss to seconds."""
    parts = time_str.strip().split(":")
    if len(parts) == 2:
        return float(parts[0]) * 60 + float(parts[1])
    elif len(parts) == 3:
        return float(parts[0]) * 3600 + float(parts[1]) * 60 + float(parts[2])
    return float(time_str)


def seconds_to_time(seconds: float) -> str:
    """Convert seconds to mm:ss.ms format."""
    m, s = divmod(seconds, 60)
    return f"{int(m)}:{s:05.2f}"


def time_to_frame(time_str: str, fps: float) -> int:
    """Convert time string to frame number."""
    return int(time_to_seconds(time_str) * fps)


def frame_to_time(frame: int, fps: float) -> str:
    """Convert frame number to time string."""
    return seconds_to_time(frame / fps)
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(VIDEO_PATH)
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

start_frame = time_to_frame(START_TIME, fps)
end_frame = time_to_frame(END_TIME, fps)

cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)
frames = []
for i in range(end_frame - start_frame):
    ret, frame = cap.read()
    if not ret:
        break
    frames.append(frame)
cap.release()

print(f"Extracted {len(frames)} frames ({len(frames)/fps:.1f}s) from {os.path.basename(VIDEO_PATH)}")
print(f"Resolution: {width}x{height}, FPS: {fps}")
```

```python
import deeplabcut

# Save frames as temp video for DLC processing
TEMP_VIDEO = "/tmp/mousercv_eval_clip.mp4"
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(TEMP_VIDEO, fourcc, fps, (width, height))
for frame in frames:
    out.write(frame)
out.release()
print(f"Saved {len(frames)} frames as temp video")

# Run SuperAnimal inference
print("Running DLC SuperAnimal-Quadruped inference...")
deeplabcut.video_inference_superanimal(
    [TEMP_VIDEO],
    superanimal_name=SUPERANIMAL_NAME,
    model_name=MODEL_NAME,
    detector_name=DETECTOR_NAME,
    video_adapt=False,  # Set True for better results (slower)
    pcutoff=PCUTOFF,
)
print("Inference complete!")
```

```python
import pandas as pd
import glob

# DLC saves results as h5 file next to the video
result_files = glob.glob(TEMP_VIDEO.replace(".mp4", "*.h5"))
if not result_files:
    # Try csv
    result_files = glob.glob(TEMP_VIDEO.replace(".mp4", "*.csv"))

print(f"Found result files: {result_files}")
df = (
    pd.read_hdf(result_files[0])
    if result_files[0].endswith(".h5")
    else pd.read_csv(result_files[0], header=[0, 1, 2], index_col=0)
)
print(f"Shape: {df.shape}")
print(f"Columns (first 10): {list(df.columns[:10])}")
print(f"\nSample data:")
df.head()
```

```python
import matplotlib.pyplot as plt

# Extract confidence scores for all keypoints
scorer = df.columns.get_level_values(0)[0]
keypoint_names = df.columns.get_level_values(1).unique()

reliability = {}
for kp in keypoint_names:
    try:
        conf = df[(scorer, kp, "likelihood")].values
        reliability[kp] = {
            "mean_conf": np.mean(conf),
            "pct_above_03": np.mean(conf > 0.3) * 100,
            "pct_above_05": np.mean(conf > 0.5) * 100,
            "pct_above_07": np.mean(conf > 0.7) * 100,
        }
    except KeyError:
        pass

rel_df = pd.DataFrame(reliability).T.sort_values("mean_conf", ascending=False)

# Color essential keypoints
colors = [
    "#22c55e" if kp in ESSENTIAL_KEYPOINTS else "#6b7280" for kp in rel_df.index
]

fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Bar chart: mean confidence per keypoint
axes[0].barh(range(len(rel_df)), rel_df["mean_conf"], color=colors)
axes[0].set_yticks(range(len(rel_df)))
axes[0].set_yticklabels(rel_df.index, fontsize=8)
axes[0].set_xlabel("Mean Confidence")
axes[0].set_title("Mean Confidence per Keypoint")
axes[0].axvline(x=0.3, color="red", linestyle="--", alpha=0.5, label="pcutoff=0.3")
axes[0].legend()
axes[0].invert_yaxis()

# Bar chart: % frames above pcutoff
axes[1].barh(range(len(rel_df)), rel_df["pct_above_03"], color=colors)
axes[1].set_yticks(range(len(rel_df)))
axes[1].set_yticklabels(rel_df.index, fontsize=8)
axes[1].set_xlabel("% Frames Above pcutoff=0.3")
axes[1].set_title("Keypoint Reliability (% frames detected)")
axes[1].axvline(x=50, color="red", linestyle="--", alpha=0.5, label="50% threshold")
axes[1].legend()
axes[1].invert_yaxis()

plt.suptitle(
    f"DLC SuperAnimal-Quadruped Keypoint Reliability\n"
    f"{os.path.basename(VIDEO_PATH)} ({START_TIME}-{END_TIME})",
    fontsize=14,
)
plt.tight_layout()
plt.show()

# Print essential keypoint summary
print("\n=== ESSENTIAL KEYPOINTS ===")
for kp in ESSENTIAL_KEYPOINTS:
    if kp in rel_df.index:
        row = rel_df.loc[kp]
        status = "RELIABLE" if row["pct_above_03"] > 50 else "UNRELIABLE"
        print(
            f"  {kp:25s} mean_conf={row['mean_conf']:.3f}  "
            f"above_0.3={row['pct_above_03']:.0f}%  {status}"
        )
    else:
        print(f"  {kp:25s} NOT FOUND in DLC output")
```

```python
# Pick 6 evenly spaced frames
sample_indices = np.linspace(0, len(frames) - 1, 6, dtype=int)

fig, axes = plt.subplots(2, 3, figsize=(20, 14))
axes = axes.flatten()

# Color map for keypoints
essential_color = (0, 255, 0)  # green
other_color = (128, 128, 128)  # gray

for ax_idx, frame_idx in enumerate(sample_indices):
    frame_rgb = cv2.cvtColor(frames[frame_idx], cv2.COLOR_BGR2RGB).copy()

    # Draw keypoints
    for kp in keypoint_names:
        try:
            x = df.iloc[frame_idx][(scorer, kp, "x")]
            y = df.iloc[frame_idx][(scorer, kp, "y")]
            conf = df.iloc[frame_idx][(scorer, kp, "likelihood")]

            if conf > PCUTOFF:
                color = essential_color if kp in ESSENTIAL_KEYPOINTS else other_color
                radius = 6 if kp in ESSENTIAL_KEYPOINTS else 3
                cv2.circle(frame_rgb, (int(x), int(y)), radius, color, -1)
                if kp in ESSENTIAL_KEYPOINTS:
                    cv2.putText(
                        frame_rgb,
                        kp.split("_")[-1],
                        (int(x) + 8, int(y) - 5),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.4,
                        color,
                        1,
                    )
        except (KeyError, IndexError):
            pass

    axes[ax_idx].imshow(frame_rgb)
    abs_frame = start_frame + frame_idx
    axes[ax_idx].set_title(
        f"Frame {abs_frame} ({frame_to_time(abs_frame, fps)})", fontsize=10
    )
    axes[ax_idx].axis("off")

plt.suptitle(
    f"Keypoint Detection — Green=Essential, Gray=Other (pcutoff={PCUTOFF})",
    fontsize=14,
)
plt.tight_layout()
plt.show()
```

```python
def compute_paw_velocities(df, scorer, fps):
    """Compute paw velocities over time.

    Returns a dict mapping paw label -> velocity array (px/sec).
    Low-confidence frames are masked as NaN.
    """
    velocities = {}

    paw_pairs = {
        "front_left_paw": "front_left",
        "front_right_paw": "front_right",
        "back_left_paw": "back_left",
        "back_right_paw": "back_right",
    }

    for kp_name, label in paw_pairs.items():
        try:
            x = df[(scorer, kp_name, "x")].values
            y = df[(scorer, kp_name, "y")].values
            conf = df[(scorer, kp_name, "likelihood")].values

            # Velocity = displacement between frames (px/frame -> px/sec)
            dx = np.diff(x)
            dy = np.diff(y)
            vel = np.sqrt(dx**2 + dy**2) * fps  # px/sec

            # Mask low-confidence frames
            mask = (conf[:-1] > PCUTOFF) & (conf[1:] > PCUTOFF)
            vel[~mask] = np.nan

            velocities[label] = vel
        except KeyError:
            print(f"Warning: keypoint {kp_name} not found")
            velocities[label] = np.full(len(df) - 1, np.nan)

    return velocities


vels = compute_paw_velocities(df, scorer, fps)

# Average front and back paw velocities
front_vel = np.nanmean([vels["front_left"], vels["front_right"]], axis=0)
back_vel = np.nanmean([vels["back_left"], vels["back_right"]], axis=0)

# Smooth with rolling window
window = int(fps * 0.5)  # 0.5 second window
front_smooth = pd.Series(front_vel).rolling(window, center=True, min_periods=1).mean()
back_smooth = pd.Series(back_vel).rolling(window, center=True, min_periods=1).mean()

# Plot
time_axis = np.arange(len(front_vel)) / fps + time_to_seconds(START_TIME)

fig, axes = plt.subplots(3, 1, figsize=(18, 10), sharex=True)

axes[0].plot(
    time_axis, front_smooth, color="#3b82f6", linewidth=1,
    label="Front paws (grooming signal)",
)
axes[0].set_ylabel("Velocity (px/s)")
axes[0].set_title("Front Paw Velocity (smoothed 0.5s) — HIGH = potential grooming")
axes[0].legend()
axes[0].grid(alpha=0.3)

axes[1].plot(
    time_axis, back_smooth, color="#f97316", linewidth=1,
    label="Back paws (scratching signal)",
)
axes[1].set_ylabel("Velocity (px/s)")
axes[1].set_title("Back Paw Velocity (smoothed 0.5s) — HIGH = potential scratching")
axes[1].legend()
axes[1].grid(alpha=0.3)

# Velocity ratio: back/front. >1 = scratching, <1 = grooming
ratio = back_smooth / (front_smooth + 1e-6)
axes[2].plot(time_axis, ratio, color="#a855f7", linewidth=1)
axes[2].axhline(y=1.0, color="gray", linestyle="--", alpha=0.5, label="ratio=1 (equal)")
axes[2].set_ylabel("Back/Front Ratio")
axes[2].set_xlabel("Time (seconds)")
axes[2].set_title("Paw Velocity Ratio — >1 = scratching, <1 = grooming")
axes[2].set_ylim(0, 5)
axes[2].legend()
axes[2].grid(alpha=0.3)

plt.suptitle(
    f"Paw Velocity Analysis — {os.path.basename(VIDEO_PATH)}", fontsize=14
)
plt.tight_layout()
plt.show()

# Summary statistics
print("\n=== PAW VELOCITY SUMMARY ===")
print(f"Front paw mean velocity: {np.nanmean(front_vel):.1f} px/s (std: {np.nanstd(front_vel):.1f})")
print(f"Back paw mean velocity:  {np.nanmean(back_vel):.1f} px/s (std: {np.nanstd(back_vel):.1f})")
pct_valid_front = np.mean(~np.isnan(front_vel)) * 100
pct_valid_back = np.mean(~np.isnan(back_vel)) * 100
print(f"Front paw valid frames: {pct_valid_front:.0f}%")
print(f"Back paw valid frames:  {pct_valid_back:.0f}%")
```

```python
# Scatter plot: front vs back paw velocity per frame
# This should show clusters if scratch/groom are separable
fig, ax = plt.subplots(figsize=(10, 10))

# Use smoothed velocities, drop NaN
mask = ~(np.isnan(front_smooth) | np.isnan(back_smooth))
sc = ax.scatter(
    front_smooth[mask], back_smooth[mask],
    alpha=0.3, s=5, c=time_axis[mask], cmap="viridis",
)
ax.plot(
    [0, ax.get_xlim()[1]], [0, ax.get_xlim()[1]],
    "r--", alpha=0.5, label="Equal velocity line",
)
ax.set_xlabel("Front Paw Velocity (px/s) — grooming axis")
ax.set_ylabel("Back Paw Velocity (px/s) — scratching axis")
ax.set_title(
    "Front vs Back Paw Velocity\n"
    "Points above line = scratching candidate\n"
    "Points below line = grooming candidate"
)
ax.legend()
ax.set_aspect("equal")
plt.colorbar(sc, label="Time (s)")
plt.tight_layout()
plt.show()
```

```python
# Track back paw y-coordinate over time
# During scratching, hind paw lifts (y decreases in image coords)
fig, axes = plt.subplots(2, 1, figsize=(18, 8), sharex=True)

for paw, color, label in [
    ("back_left_paw", "#f97316", "Left hind"),
    ("back_right_paw", "#ef4444", "Right hind"),
]:
    try:
        y = df[(scorer, paw, "y")].values
        conf = df[(scorer, paw, "likelihood")].values
        y_masked = np.where(conf > PCUTOFF, y, np.nan)

        time_ax = np.arange(len(y)) / fps + time_to_seconds(START_TIME)
        axes[0].plot(
            time_ax, y_masked, color=color, linewidth=0.5, alpha=0.7, label=label
        )
    except KeyError:
        pass

axes[0].set_ylabel("Y position (pixels, 0=top)")
axes[0].set_title("Hind Paw Y Position — Dips = paw lifts (potential scratching)")
axes[0].legend()
axes[0].invert_yaxis()  # Lower y = higher in image
axes[0].grid(alpha=0.3)

# Also show back paw velocity with threshold line
axes[1].plot(time_axis, back_smooth, color="#f97316", linewidth=1)
threshold_90 = np.nanpercentile(back_smooth, 90)
axes[1].axhline(
    y=threshold_90, color="red", linestyle="--", alpha=0.7,
    label=f"90th percentile ({threshold_90:.0f} px/s)",
)
axes[1].set_ylabel("Velocity (px/s)")
axes[1].set_xlabel("Time (seconds)")
axes[1].set_title("Back Paw Velocity — peaks above threshold = scratch bout candidates")
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.suptitle(
    f"Hind Paw Analysis — {os.path.basename(VIDEO_PATH)}", fontsize=14
)
plt.tight_layout()
plt.show()
```
